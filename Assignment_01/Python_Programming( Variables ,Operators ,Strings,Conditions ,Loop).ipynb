```python
# Age Checker
age = int(input("Enter your age: "))

if age < 0:
    print("Invalid age")
elif age > 18:
    print("You are an adult")
elif 13 <= age <= 17:
    print("You are a teenager")
else:
    print("You are a child")
```

```python
#Grade Calculator
mark=int(input("Enter your mark:"))

if mark < 0 or mark > 100:
  result="Invalid mark"
elif mark >= 90 and mark <= 100:
  result="A+"
elif mark >= 80 and mark <= 89:
  result="A"
elif mark >= 70 and mark <= 79:
  result="B"
elif mark >= 60 and mark <= 69:
  result="C"
elif mark >= 50 and mark <= 59:
  result="D"
else:
  result="F __Failed"

print(f"Your grade is {result}")
```

```python
# Login System
correct_username = "admin"
correct_password = "1234"

attempts = 0
max_attempts = 3 #counter

while attempts < max_attempts:
    username = input("Enter username: ")
    password = input("Enter password: ")

    if username == correct_username and password == correct_password:
        print(f"Login successful! Welcome, {username}.")
        break
    elif username == correct_username and password != correct_password:
        print("Incorrect password. Try again.")
    else:
        print("User not found.")

    attempts = attempts+1

if attempts == max_attempts:
    print("Account locked.")

```

**Question 5: ATM Simulation**

```python
#an ATM program with the following variables: pinNumber, balance, and request:
pinNumber = 1111
balance = 5000

pin = int(input("Enter your PIN: "))

if pin == pinNumber:
    request = int(input("Enter withdrawal amount: "))

    if request > 500 and request <= balance:
        balance = balance - request
        print("Withdrawal successful.")
        print(f"Remaining balance:{balance}")
    elif request > balance:
        print("Insufficient balance.")
    else:
        print("Withdrawal amount must be greater than 500.")
else:
    print("Wrong PIN.")
```

```python
#an attempts counter
pinNumber = 1111
balance = 5000
attempts = 3 #counter

while attempts > 0:
    pin = int(input("Enter your PIN: "))

    if pin == pinNumber:
        request = int(input("Enter withdrawal amount: "))

        if request > 500 and request <= balance:
            balance = balance - request
            print("Withdrawal successful.")
            print(f"Remaining balance: {balance}")
        elif request > balance:
            print("Insufficient balance.")
        else:
            print("Withdrawal amount must be greater than 500.")
        break
    else:
        attempts = attempts - 1
        print(f"Wrong pin. Attempts remaining: {attempts}")

        if attempts == 0:
            print("Card blocked.")
```

**Question 6: Basic Loops**

**For loop**

```python
#Print numbers from 1 to 20
for i in range(1, 21):
    print(i)
```

```python
#Multiplication table of 7
for i in range(1, 11):
    print(f"7 x {i} = {7 * i}")
```

```python
#Sum of numbers from 1 to 100
total = 0
for i in range(1, 101):
    total = total + i
print(f"Sum of numbers from 1 to 100: {total}")
```

```python
#Even numbers between 1 and 50
for i in range(2, 51, 2):
    print(i)
```

```python
#Fruits list with position
fruits = ['apple', 'banana', 'cherry', 'mango', 'grape']
for index, fruit in enumerate(fruits, start=1):
    print(index, fruit)
```

**While loop**

```python
#Print numbers from 10 down to 1
i = 10
while i >= 1:
    print(i)
    i = i - 1
```

```python
#Sum of positive numbers until 0 or negative is entered
total = 0
while True:
    num = int(input("Enter a number: "))
    if num <= 0:
        break
    total = total + num
print(f"The total sum of all numbers: {total}")
```

```python
#Countdown with Blast off
count = 10
while count >= 0:
    if count == 0:
        print("Blast off!")
    else:
        print(count)
    count = count - 1
```

```python
#First 10 multiples of 6
i = 1
while i <= 10:
    print(6 * i)
    i = i + 1
```
